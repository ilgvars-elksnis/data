```python
df1 = pd.DataFrame(data_frame, columns=['county', 'cases', 'deaths'])
```

```python
df1
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='cases', y='deaths', data=df1)
graph
```

```python
df2 = df1[(df1.county == 'New York City') & (df1.cases >= 500)]
df2
```

```python
g = sns.relplot(x="cases", y="deaths", kind="line", data=df2)
g.fig.autofmt_xdate()
```
